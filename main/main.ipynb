```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
reg = linear_model.LinearRegression()
```

```python
df = pd.read_csv("ipl_200plus_scores.csv")
```

```python
year = df.Season
scores = df["200+ Scores"]
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['Season']] - 2008, scores);
```

```python
print(reg.predict(pd.DataFrame([[12]], columns=["Season"])))
print(reg.coef_)
print(reg.intercept_)
```

```python
plt.xlabel("Years")
plt.ylabel("200+ scores")

years = df["Season"] - 2008
score = df["200+ Scores"]

plt.scatter(years, score)

plt.plot(years, reg.coef_ * years + reg.intercept_, color="orange")
```
